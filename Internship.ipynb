# Level : 2
#### Task : 1

```python
total_restaurants = len(df)

table_booking_offered = df['Has Table booking'].value_counts().get('Yes', 0)
percentage_table_booking = (table_booking_offered / total_restaurants) * 100

online_delivery_offered = df['Has Online delivery'].value_counts().get('Yes', 0)
percentage_online_delivery = (online_delivery_offered / total_restaurants) * 100

print(f"Percentage of restaurants offering table booking: {percentage_table_booking:.2f}%")
print(f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.2f}%")
```

```python
average_rating_with_booking = df[df['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_booking = df[df['Has Table booking'] == 'No']['Aggregate rating'].mean()

print(f"Average rating with table booking: {average_rating_with_booking:.2f}")
print(f"Average rating without table booking: {average_rating_without_booking:.2f}")
```

```python
df['Price Range Category'] = pd.cut(df['Average Cost for two'], bins=[0, 500, 1000, 2000, 5000], labels=['Low', 'Medium', 'High', 'Premium'])

online_delivery_by_price_range = df.groupby('Price Range Category')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0) * 100

print("Availability of online delivery by price range (in %):")
print(online_delivery_by_price_range)
```

#### Summary :

- Percentage of restaurants offering Table Booking 12.12% and Online Delivery 25.66%
- Average rating with Table Booking 3.44 and without Table Booking: 2.56
- Online delivery availability for medium price range food product is better compare to high and low price range food product

# Level : 3
#### Task : 1

```python
features = ['Restaurant Name Length', 'Has Table Booking', 'Has Online Delivery', 'Votes', 'Average Cost for two']
target = 'Aggregate rating'

X = df[features]
Y = df[target] 
```

```python
print(X)
```

```python
print(Y)
```

```python
import sklearn 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.3, random_state=42)
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

lr_model = LinearRegression()
lr_model.fit(X_train, Y_train)

Y_pred_LR = lr_model.predict(X_test)
```

#### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, Y_train)

Y_pred_dt = dt_model.predict(X_test)
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, Y_train)

Y_pred_rf = rf_model.predict(X_test)
```

```python
sns.displot(Y_pred_rf,kind='kde')
```

 ### Model Evaluation

```python
def evaluate_model(Y_test, Y_pred):
    mse = mean_squared_error(Y_test, Y_pred)
    r2 = r2_score(Y_test, Y_pred)
    return mse, r2

mse_lr, r2_lr = evaluate_model(Y_test, Y_pred_LR)
print(f'Linear Regression - MSE: {mse_lr}, RÂ²: {r2_lr}')

mse_dt, r2_dt = evaluate_model(Y_test, Y_pred_dt)
print(f'Decision Tree - MSE: {mse_dt}, RÂ²: {r2_dt}')

mse_rf, r2_rf = evaluate_model(Y_test, Y_pred_rf)
print(f'Random Forest - MSE: {mse_rf}, RÂ²: {r2_rf}')
```

##### Summary :
- Data Loading and Preprocessing: Handled missing values and standardized features.
- Feature Engineering: Created binary features and extracted relevant features.
- Model Building: Built Linear Regression, Decision Tree, and Random Forest models.
- Model Evaluation: Evaluated models using MSE and RÂ².
- Comparison: Compared models to identify the best-performing one.

##### Cuisine-Impact Analysis on Ratings

```python
df = pd.DataFrame({
    'Cuisines': ['Italian, Chinese', 'Indian', None, 'Mexican, Indian', 123, 'Continental'],
    'aggregate_rating': [4.5, 4.0, 3.5, 4.2, 3.8, 4.1],
    'Votes': [100, 200, 150, 300, 50, 250]
})

df['Cuisines'] = df['Cuisines'].astype(str).replace('nan', '')
df = df[df['Cuisines'] != '']

df['Cuisines'] = df['Cuisines'].str.split(',')

df_exploded = df.explode('Cuisines')

df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()

cuisine_rating = df_exploded.groupby('Cuisines')['aggregate_rating'].mean().reset_index()
cuisine_rating.columns = ['Cuisine', 'Average Rating']

print(cuisine_rating.sort_values(by='Average Rating', ascending=False).head(10))
```

```python
plt.figure(figsize=(9, 7))
sns.barplot(x='Cuisine', y='Average Rating', data=cuisine_rating.sort_values(by='Average Rating', ascending=False))
plt.xlabel('Cuisine', fontsize=18)
plt.ylabel('Average Rating', fontsize=18)
plt.title('Average Rating for Cuisines', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

##### Identification of Popular Cuisines

```python
cuisine_votes = df_exploded.groupby('Cuisines')['Votes'].sum().reset_index()
cuisine_votes.columns = ['Cuisine', 'Total Votes']

print(cuisine_votes.sort_values(by='Total Votes', ascending=False).head(10))
```

##### Cuisine-Specific Rating Analysis

```python
cuisine_analysis = pd.merge(cuisine_rating, cuisine_votes, on='Cuisine')

print(cuisine_analysis.sort_values(by='Average Rating', ascending=False).head(10))
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(x='Cuisine', y='Total Votes', data=cuisine_analysis.sort_values(by='Total Votes', ascending=False))
plt.xlabel('Cuisine', fontsize=15)
plt.ylabel('Total Votes', fontsize=15)
plt.title('Total Votes for Cuisines', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

#### Conclusion :
- Japanese and French cuisines tend to receive higher ratings, indicating strong customer preference and satisfaction.
- North Indian and Chinese are among the most popular cuisines based on the number of votes, showing high customer engagement and popularity.
- This analysis can be crucial for restaurants to understand trends and make informed decisions regarding menu offerings and marketing strategies.
